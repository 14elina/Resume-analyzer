```python
import pdfplumber
def extract_text(file):
    text = ""
    with pdfplumber.open(file) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

file = r"C:\Users\PC\Documents\resume.pdf" 
text = extract_text(file)

skills = ["Python", "Data Analysis", "Machine Learning", "Natural Language Processing", "Web Development", "Database Management", "Cloud Computing", "Django", "Cybersecurity", "Project Management"]

found_skills = []

for skill in skills:
    if skill.lower() in text.lower():
        found_skills.append(skill)
        
        
print("Skills found:")
print(found_skills)
```
